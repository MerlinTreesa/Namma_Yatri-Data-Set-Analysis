Data Preparation

```python
df['Searches'].replace(',','',regex=True,inplace=True)
df['Searches which got estimate'].replace(',','',regex=True,inplace=True)
df['Searches for Quotes'].replace(',','',regex=True,inplace=True)
df['Searches which got Quotes'].replace(',','',regex=True,inplace=True)
df['Search-to-estimate Rate'].replace('%','',regex=True,inplace=True)
df['Rider Fare Acceptance  Rate'].replace('%','',regex=True,inplace=True)
df['Driver Quote Acceptance Rate'].replace('%','',regex=True,inplace=True)
df['Quote-to-booking Rate'].replace('%','',regex=True,inplace=True)
df['Booking Cancellation Rate'].replace('%','',regex=True,inplace=True)
df['Driver Cancellation Rate'].replace('%','',regex=True,inplace=True)
df['User Cancellation Rate'].replace('%','',regex=True,inplace=True)
df['Conversion Rate'].replace('%','',regex=True,inplace=True)

df["Drivers' Earnings"].replace('₹','',regex=True,inplace=True)
df["Drivers' Earnings"].replace(',','',regex=True,inplace=True)
df['Average Fare per Trip'].replace('₹','',regex=True,inplace=True)
df['Average Fare per Trip'].replace(',','',regex=True,inplace=True)
```

```python
df['Searches which got Quotes'].replace(',','',regex=True,inplace=True)
df['Bookings'].replace(',','',regex=True,inplace=True)
df['Cancelled Bookings'].replace(',','',regex=True,inplace=True)
df['Completed Trips'].replace(',','',regex=True,inplace=True)
df['Distance Travelled (km)'].replace(',','',regex=True,inplace=True)
```

```python
object_cols = df.select_dtypes(include=['object']).columns
df[object_cols] = df[object_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.rename(columns={'Search-to-estimate Rate':'Search-to-estimate Rate (%)',
                   'Rider Fare Acceptance  Rate':'Rider Fare Acceptance  Rate (%)',
                  'Driver Quote Acceptance Rate':'Driver Quote Acceptance Rate (%)',
                  'Quote-to-booking Rate':'Quote-to-booking Rate (%)',
                  'Booking Cancellation Rate':'Booking Cancellation Rate (%)',
                  'Driver Cancellation Rate':'Driver Cancellation Rate (%)',
                  'User Cancellation Rate':'User Cancellation Rate(%)',
                  'Conversion Rate':'Conversion Rate (%)',
                  "Drivers' Earnings":"Drivers' Earnings (Rs)",
                  'Average Fare per Trip':'Average Fare per Trip (Rs)'}, inplace=True)


```

Feature Engineering

```python
df['Driver Earnings per Trip (Rs)'] = round(df["Drivers' Earnings (Rs)"] / df['Completed Trips'])
```

```python
df['Driver Earnings per KM (Rs)'] = round(df["Drivers' Earnings (Rs)"] / df['Distance Travelled (km)'])
```

```python
df=df.drop(index=8)
```

```python
for i in df.select_dtypes(include='number').columns:
  plt.boxplot(df[i])
  plt.title(i)
  plt.show()
```

Exploratory Data Analysis

```python
for i in df.columns:
    if df[i].dtypes != 'object':
        city_data = df.groupby('City')[i].mean()

        plt.figure(figsize=(6, 4))
        plt.title(f"City-wise {i}")
        bars = plt.bar(x=city_data.index, height=city_data.values)
        # Add labels on the bars
        for bar in bars:
            height = bar.get_height()
            plt.text(
                bar.get_x() + bar.get_width() / 2,
                height,f'{height:.2f}',
                ha='center', va='bottom'
            )
        plt.xticks(rotation=90)
        plt.xlabel('City')
        plt.ylabel(i)
        plt.show()
```
